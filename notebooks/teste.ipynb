## Concatenar records

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
records = ['426627000','426783006']
```

```python
df = pd.read_csv('../records/data_beat_'+records[0]+'.csv')
df2 = pd.read_csv('../records/data_beat_'+records[1]+'.csv')
```

```python
df
```

```python
df = df.drop(columns=['Unnamed: 0'])
```

```python
plt.figure(figsize=(15,5))
plt.plot(df.loc[4, :'650'])
```

```python
df2
```

```python
df2 = df2.drop(columns=['Unnamed: 0'])
```

```python
plt.figure(figsize=(15,5))
plt.plot(df2.loc[4, :'650'])
```

```python
df_concat = pd.concat([df,df2])
```

```python
df_concat
```

```python
import numpy as np
```

```python
plt.figure(figsize=(15,5))
plt.plot(np.linspace(0,650,651),df_concat.loc[4,:'650'].iloc[0])
```

```python
df_concat.to_csv("batimentos.csv")
```
