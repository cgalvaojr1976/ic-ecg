## Concatenar records

Notebook para concatenar diferentes batimentos

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#records = ['426627000','426783006']
records = ['164947007','426783006']
```

```python
df = pd.read_csv('../beats/data_beat_'+records[0]+'.csv')
df2 = pd.read_csv('../beats/data_beat_'+records[1]+'.csv')
```

```python
df
```

```python
df = df.drop(columns=['Unnamed: 0'])
```

```python
df2 = df2.drop(columns=['Unnamed: 0'])
```

```python
df_concat = pd.concat([df,df2])
```

```python
df_concat
```

```python
df_concat.to_csv("../models/batimentos.csv")
```
