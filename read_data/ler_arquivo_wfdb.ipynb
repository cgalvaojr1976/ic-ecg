```python
filename = 'A0001'
#filename = 'Q0001'
#filename = 'I01'
#filename = 's0010_re'
#filename = '00002_hr'
#filename = 'E00002'
#filename = 'JS00001'
#filename = 'JS10647'
record = wfdb.rdrecord(PATH+filename) 
```

```python
print(record.__dict__['n_sig'])
print(record.__dict__['fs'])
display(record.__dict__['comments'][2])
```

```python
wfdb.plot_wfdb(record=record, title='Record a103l from PhysioNet Challenge 2015') 
```

```python
Ncols = 2
Nrows = np.ceil(record.__dict__['n_sig']/2).astype('int32')

fig, axs = plt.subplots(Nrows,Ncols,figsize=(25,15),sharex=True)
N = len(record.__dict__['p_signal'][:,0])
time = np.linspace(0,N-1,N)/record.__dict__['fs']
for i in range(record.__dict__['n_sig']):
    xx = i//Ncols
    yy = i%Ncols
    axs[xx,yy].plot(time, record.__dict__['p_signal'][:,i])
    axs[xx,yy].set_title('Lead '+record.__dict__['sig_name'][i])
    axs[xx,yy].set_ylabel('ECG ('+record.__dict__['units'][i]+')')
    axs[xx,yy].axis([0,max(time), -1, 1])
for i in range(Ncols):
    axs[xx,i].set_xlabel('time (s)')
```
