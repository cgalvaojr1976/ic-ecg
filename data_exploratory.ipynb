Primeiramente verificar como estão os fields em cada um dos bancos de dados. Para isso, vamos usar um arquivo aleatório de cada banco

```python
paths = ["WFDB_ChapmanShaoxing/", "WFDB_CPSC2018/", "WFDB_CPSC2018_2/",\
    "WFDB_Ningbo/", "WFDB_PTB/", "WFDB_PTBXL/", "WFDB_StPetersburg/"]
```

Erro no banco de dados 3 "WFDB_Ga/"

```python
files = ["JS09339", "A5504", "Q3220", "JS41191", \
    "S0493", "HR19110", "I0061"]
```

```python
# Reconhecer os dados categóricos dos bancos

fields = []
for x in range(len(paths)):
    record, field = wfdb.rdsamp(paths[x]+files[x], channels=[0])
    fields.append(field)
    
fields
```

```python
# Transformar os comentários em Dicionários

for y in range(len(fields)):
    for x in range(len(fields[y]['comments'])):
        fields[y][fields[y]['comments'][x].split(": ")[0]] = fields[y]['comments'][x].split(": ")[1]

fields
```

```python
# Criar tabela

df = pd.DataFrame(fields, index = files)
df = df.drop('comments', 1)

df
```

```python
paths2 = ["WFDB_ChapmanShaoxing", "WFDB_CPSC2018", "WFDB_CPSC2018_2",\
    "WFDB_Ningbo", "WFDB_PTB", "WFDB_PTBXL", "WFDB_StPetersburg"]
```

```python
lala = []
nomes = []
gravacoes = []
problemas = 0

for diretorios in paths2:
    for root, dirs, files in os.walk(diretorios, topdown=False):
        for name in files:
            temp = name.split('.')
#           print(root)
#           print(temp)
            if len(temp)==2:
                if name.split('.')[1] == 'hea':
                    arquivo = os.path.join(root, temp[0])
#                    print(arquivo)
                    try:
                        record, field = wfdb.rdsamp(arquivo, channels=[1])
                        nomes.append(nomes.split(.)[1])
                        gravacoes.append(record)
                        lala.append(field)
                    except:
                        problemas += 1
```

```python
print(problemas)
```

```python
for y in range(len(lala)):
    for x in range(len(lala[y]['comments'])):
        lala[y][lala[y]['comments'][x].split(": ")[0]] = lala[y]['comments'][x].split(": ")[1]
```

```python
lala
```

```python
df2 = pd.DataFrame(lala)
df2 = df2.drop('comments', 1)

df2
```

```python
disaes = [df2["Dx"][x] for x in range(len(df2["Dx"]))]
disaes = [disaes[x].split(',') for x in range(len(disaes))]
```

```python
# Transformar os disaes em Inteiros
disae_f =[]
for x in range(len(disaes)):
    # Caso seja um vetor de disaes
    if (len(disaes[x]) > 1):
        for y in range(len(disaes[x])):
            disae_f.append(int(disaes[x][y]))
    else:
        disae_f.append(int(disaes[x][0]))
```
