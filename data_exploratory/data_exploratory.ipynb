```python
# Transformar os disaes em Inteiros
disae_f =[]
for x in range(len(disaes)):
    # Caso seja um vetor de disaes
    if (len(disaes[x]) > 1):
        for y in range(len(disaes[x])):
            disae_f.append(int(disaes[x][y]))
    else:
        disae_f.append(int(disaes[x][0]))
```
